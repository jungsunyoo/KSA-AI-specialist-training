# 웹 API로 데이터 추출하기

```python
import requests
import json
# API 키 지정
apikey = "3880df9eec35bb4a06a95618dba80b2e"
# 날씨를 확인할 도시 지정하기
cities=["Seoul,KR", "Tokyo, JP", "New York, US"]

session=requests.session()
# API 지정
api="http://api.openweathermap.org/data/2.5/forecast?q={city}&APPID={key}"

k2c = lambda k: k - 273.15

# 각 도시의 정보 추출하기
for name in cities: 
    url = api.format(city=name, key=apikey)
    # API에 요청을 보내 데이터 추출하기
    r = requests.get(url)
#     print(r.text)
    # 결과를 JSON 형식으로 변환하기
    data=json.loads(r.text)
#     print(data["cod"],"\n")
#     print(data["list"])
    print("+ 도시 =", data["city"]["name"])
    print("| 최저 기온 = ", k2c(data["list"][0]["main"]["temp_min"]))
    print("| 최고 기온 = ", k2c(data["list"][0]["main"]["temp_max"]))
    print("| 습도 =", data["list"][0]["main"]["humidity"])
#     print("| 날씨 =", data["list"][0]["weather"]["description"])
    print("| 기압 =", data["list"][0]["main"]["pressure"])
    print("| 풍향 =", data["list"][0]["wind"]["deg"])
    print("| 풍속 =", data["list"][0]["wind"]["speed"])
    print("")
#     print("+ 도시 =", data['name'])
```

```python
import os
import sys
import datetime
import time
import urllib.request
import json

# 네이버 api용 함수
def get_request_url(url):
    client_id="cGhf645hHp2vT5iIp0x4"
    client_secret="XkYP1ZBdhH"
    rep=urllib.request.Request(url)
    rep.add_header("X-Naver-Client-Id",client_id)
    rep.add_header("X-Naver-Client-Secret",client_secret)
    try: 
        response=urllib.request.urlopen(rep)
        if response.getcode()==200: # 정상적으로 처리 됨
            print("[%s] Url request success" % datetime.datetime.now())
            return response.read().decode("utf-8")
    except Exception as e: # 실패했을 때
        print(e)
        print("[%s] Error for URL : $s" % datetime.datetime.now(), url)
        return None # 에러 났으니까 아무것도 출력 안하게
    
def getNaverSearchResult(sNode, search_text, page_start, display) :
    base="https://openapi.naver.com/v1/search"
    node='/blog.json' # 뉴스 카페 블로그 등. 확장자 이름 (json)
    params="?query=%s&start=%s&display=%s" % (urllib.parse.quote(search_text), page_start, display)
    url = base + node + params
    getData = get_request_url(url)
    if getData == None : 
        return None
    else : 
        return json.loads(getData)
        
def getPostData(post, jsonResult) : 
    title = post["title"]
    description=post["description"]
    link=post["link"]
    org_link = post["bloggerlink"]
#     datetime.datetime.strptime(timeStr, '%Y-%m-%d
#     year = math.floor(int(post["postdate"])/10**4)
#     month= math.floor((int(post["postdate"])-year*10000)/10**2)
#     day = math.floor(year*10000)
#     pDate=datetime.date(math.floor(int(post["postdate"])/10**4), 
#                         math.floor((post["postdate"]-math.floor(int(post["postdate"])/10**4))/10**2)    )
    pDate=datetime.datetime.strptime(post["postdate"], '%a, %d %b %Y %H:%M:%S +0900')
    pDate=pDate.strftime("%Y-%m-%d %H:%M:%S")
    jsonResult.append({"title":title, "description":description, "org_link":org_link, "link":link, "pDate":pDate})
    
    return

def main(): # 메인함수
    jsonResult=[] # 초기화
    sNode="news" # or "blog", "cafearticle"
    search_text="빅데이터" # 검색할 단어
    display_count=100
    page_start=1
    jsonSearch=getNaverSearchResult(sNode, search_text, page_start, display_count)
    while ((jsonSearch != None) and (jsonSearch["display"] != 0)):
        for post in jsonSearch["items"]: # post 하나씩 가져오기
            getPostData(post, jsonResult)
        nStart=jsonSearch["start"]+jsonSearch["display"] # json 다음페이지 시작번호 설정 (시작번호 + 한 화면에 보여지는 숫자)
        jsonSearch=getNaverSearchResult(sNode, search_text, nStart, display_count) #나머지는 동일하지만 시작번호가 업데이트됨
        
    with open("%s_naver_$s.json" % (search_text, sNode), "w", encoding = "utf-8") as outfile:
        retJson=json.dumps(jsonResult, indent=4, sort_keys=True, ensure_ascii=False) # indent는 들여쓰기 (가독성을 위해)
        outfile.write(retJson)
    print("$s_naver_$s.json SAVED", (search_txt, sNode))
    
if __name__=="__main__":
    main()
                     
main()
```

# 웹의 다양한 데이터 형식

```python
import openpyxl
```

```python
filename="C:/Users/Affinity/Desktop/modeule04/ch01/stats_104102.xlsx"
book=openpyxl.load_workbook(filename)
# 맨 앞의 시트 추출하기
sheet=book.worksheets[0]
# 시트의 각 행을 순서대로 추출하기
data=[]
for row in sheet.rows:
    data.append([
        row[0].value,
        row[9].value
    ])
    
# 필요없는 줄 (헤더, 연도, 계) 제거하기
del data[0]
del data[1]
del data[2]

# 데이터를 인구 순서로 정렬합니다.
data = sorted(data,key=lambda x:x[1])

# 하위 5위를 출력합니다.
for i, a in enumerate(data):
    if (i>=5): break
    print(i+1, a[0], int(a[1]))
```
