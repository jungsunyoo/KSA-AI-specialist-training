### 매개변수를 추가해 요청을 전송하는 방법

```python
# 데이터 다운로드 하기
import urllib.parse
API='http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp'
# 매개변수 (딕셔너리 자료형)를 URL 인코딩
value={'stnId':'184'} # 원하는 지역에 따라 숫자 바꾸면 됨
params=urllib.parse.urlencode(value)
print(params)
# 요청 URL 생성
url=API+"?"+params # url 리소스를 열기
print("url={0}".format(url))
```

### for문 이용

```python
import sys
import urllib.request as req
import urllib.parse as parse

# 입력줄 매개변수 추출
text=[] # 데이터 저장할 리스트
while(True):  # 무한루프
    regionNumber=input("USAGE: download-forecast-argv :")
    
    #반복구분 종료 조건
    if(regionNumber.upper()=='EXIT'): # 소문자일수도 있으니까 대문자로 바꿔주기
        break
    elif (int(regionNumber) not in [108, 109, 105, 131, 146, 156, 134, 159, 184]) :
        continue
        
    # 매개변수를 url 인코딩 (한글을 포함하는 경우 필수적으로 실행
    API='http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp'
    
    values={"stnId":regionNumber}
    params=parse.urlencode(values)
    url=API+"?"+params
    print("URL=", url)
    
    # 페이지 다운로드
    data=req.urlopen(url).read()
    text.append(data.decode('utf-8'))
```

# Beautiful Soup으로 스크레이핑하기

```python
from bs4 import BeautifulSoup

# 분석할 html
html="""
<html><body>
<h1>스크레이핑이란?</h1>
<p>웹 페이지를 분석하는 것</p>
<p>원하는 부분을 추출하는 것</p>
</body></html>


"""

# HTML 분석하기
soup=BeautifulSoup(html, "html.parser") # Beautiful 인스턴스 생성

#원하는부분 추출
h1=soup.html.body.h1
p1=soup.html.body.p
p2=p1.next_sibling.next_sibling

# 요소의 글자 출력
print("h1={0}".format(h1.string))
print("p1={0}".format(p1.string))
print("p2={0}".format(p2.string))
```

```python
from bs4 import BeautifulSoup

# fp=open('C:/Users/Affinity/Desktop/modeule04/ch01/HTML_Exam.html', "r", encoding)
HTML="""
<html>
    <body>
        <h1>스크레이핑이란?</h1><p>웹 페이지를 분석하는 것</p>
<p>원하는 부분을 추출하는 것</p>
    </body>
</html>



"""
soup=BeautifulSoup(HTML, 'html.parser')
print(HTML)
print(soup)
print(soup.prettify()) # 이건 자식/부모태그 들여쓰기 된 상태로 출력
Tit=soup.html.body.p # 특정한 요소만 뽑을 수도 있다. (부모 태그.자식 태그) 
# p 두개인데 그냥 p라고 부르면 첫번째것만 반환한다. 두 번째 p 반환하고 싶으면 Tit=soup.html.body.p.next_sibling으로 해야함
Tit2=soup.body.p.previous_sibling # 일렬로 붙어있어야만 인식됨,왜냐하면 줄바꿈 문자를 문자로 인식하기 때문에
# 따라서 사이에 줄바꿈 되어있다면 next_sibling 두번 써야함
print(Tit)
print(Tit2)
print("이전 태그", Tit2.string, "줄 바꾸기")
```

## CSS 선택자

```python
# 파일 열기
fp=open('C:/Users/Affinity/Desktop/modeule04/ch01/CSS_Exam.html', 'r', encoding='UTF8')
soup=BeautifulSoup(fp, 'html.parser')

# CSS 쿼리로 추출하기
# 제목 부분 추출하기
h1=soup.select_one('div#meigen > h1').string
Tag=soup.find('div', id='meigen').h1 #이렇게 해도 된다!
print('h1={0}'.format(h1)) 
print('h1={0}'.format(Tag.string)) 

# 목록 부분 추출
li_list=soup.select('div#meigen > ul.items > li')

print("\nli_list={0}\n".format(li_list))

for li in li_list:
    print("li={0}".format(li.string))
fp.close()
```

```python
url='https://finance.naver.com/marketindex/'
res=req.urlopen(url)

soup=BeautifulSoup(res, 'html.parser')
#제목 부분 추ㅜㄹ하기
h2=soup.select_one('#container > div.market_include > div > div.market1 > div.title > h2 > span').string # copy selector해서 붙임
print("***{0}***".format(h2))

title=soup.select_one('h3.h_lst > span.blind').string
val=soup.select_one('div.head_info > span.value').string

print("{0}={1}\n".format(title,val))

title_list=soup.select('h3.h_lst > span.blind')
val_list=soup.select('div.head_info > span.value')
n=len(title_list)
for i in range(0,n):
    print("{0} : {1}".format(title_list[i].string, val_list[i].string))
```
